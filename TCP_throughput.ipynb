**Throughput and CWnd tests**

There are bassically two kinds of figures generated in this python file

1. 20s throughput with 0.1s interval

2. 10s throughput with 1s interval

The first figure is too dense and not good for identification. Therefore, I created a much general one with less data.

**Read json files**

```python
import json
import numpy as np

#Read reno wried throughput json file
with open("data/reno_delay.json", 'r') as reno_file:
    reno_data = json.load(reno_file)
    
intervals = reno_data['intervals']
throughput_reno_wired = []
cwnd_reno_wired =[]


for interval in intervals:
    throughput = interval['streams'][0]['bits_per_second']
    window_size = interval['streams'][0]['snd_cwnd']
    throughput_reno_wired.append(throughput/1000000)
    cwnd_reno_wired.append(window_size/1000000)

# print(throughput_reno_wired)
# print(cwnd_reno_wired)
```

**Throughput and congestion window figures**

```python
import matplotlib.pyplot as plt
   
time =  np.linspace(0,20,200)

plt.plot(time, throughput_reno_wired, "r-", label = 'Reno wired')
 
plt.plot(time, throughput_reno_wireless, "r--", label = 'Reno wireless')

plt.plot(time, throughput_cubic_wired, "b-", label = 'Cubic wired')

plt.plot(time, throughput_cubic_wireless, "b--", label = 'Cubic wireless')

plt.plot(time, throughput_vegas_wired, "c-", label = 'Vegas wired')

plt.plot(time, throughput_vegas_wireless , "c--", label = 'Vegas wireless')

plt.plot(time, throughput_bbr_wired, "g-", label = 'BBR wired')

plt.plot(time, throughput_bbr_wireless, "g--", label = 'BBR wireless')

plt.title('Throughput of wired and wireless connections', fontsize=14)
plt.xlabel('Time(s)', fontsize=14)
plt.ylabel('Throughput(Mbits/sec)', fontsize=14)
plt.legend(loc=2, bbox_to_anchor=(1.05, 1.0), borderaxespad = 0.)
plt.grid(True)
plt.show()
```

```python
plt.plot(time, cwnd_reno_wired, "g-", label = 'Reno wired')
plt.plot(time, cwnd_reno_wireless, "g--", label = 'Reno wireless')
plt.plot(time, cwnd_cubic_wired, "c-", label = 'Cubic wired')
plt.plot(time, cwnd_cubic_wired, "c--", label = 'Cubic wireless')
plt.plot(time, cwnd_vegas_wired, "b-", label = 'Vegas wired')
plt.plot(time, cwnd_vegas_wired, "b--", label = 'Vegas wireless')
plt.plot(time, cwnd_bbr_wired, "r-", label = 'BBR wired')
plt.plot(time, cwnd_bbr_wired, "r--", label = 'BBR wireless')

plt.title('CWnd of all variances change over time', fontsize=14)
plt.xlabel('Time(s)', fontsize=14)
plt.ylabel('CWnd(MBytes)', fontsize=14)

plt.legend(loc=2, bbox_to_anchor=(1.05, 1.0), borderaxespad = 0.)
#plt.grid(True)
plt.show()
```
